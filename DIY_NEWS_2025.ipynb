## kota

ganti jadi lainnya aja

```python
df['kota'] = df['kota'].fillna('lainnya')
```

```python
def handle_dict_string(x):
    # if isinstance(x, str) and x.startswith('{') and x.endswith('}'):
    if isinstance(x, str) and x.startswith('[{') and x.endswith('}]'):
        return 'lainnya'
    return x

df['kota'] = df['kota'].apply(handle_dict_string)
```

```python
import re

def clean_list_string(x):
    # cek kalau berbentuk list string
    if isinstance(x, str) and (x.startswith('[') and x.endswith(']')):
        # hapus tanda [ ] dan ' , lalu strip spasi
        x_clean = re.sub(r"[\[\]']", '', x).strip()
        return x_clean
    return x

df['kota'] = df['kota'].apply(clean_list_string)
```

```python
df['kota'] = df['kota'].str.replace(r"^(kota|kabupaten|provinsi)\s+", "", regex=True)
```

```python
df['kota'] = df['kota'].replace("{'authors': None}", 'lainnya')
```

## provinsi

```python
import pandas as pd
import numpy as np
import ast

# fungsi ekstrak lat/lon, sisanya jadi NaN
def extract_lat_lon(x):
    if isinstance(x, str) and x.startswith('[{'):
        try:
            coords = ast.literal_eval(x)
            return coords[0]['lat'], coords[0]['lon']
        except:
            return np.nan, np.nan
    else:
        return np.nan, np.nan

# terapkan fungsi
df[['lat', 'lon']] = df['provinsi'].apply(lambda x: pd.Series(extract_lat_lon(x)))
```

## konversi tanggal ke datetime publishesd

```python
import pandas as pd

# ubah ke datetime
df['published'] = pd.to_datetime(df['published'], errors='coerce', utc=True)

# buat kolom kategori jam
# jika mau ke WIB (+7)
df['published_local'] = df['published'].dt.tz_convert('Asia/Jakarta')

def kategori_jam(hour):
    if 5 <= hour < 11:
        return 'pagi'
    elif 11 <= hour < 15:
        return 'siang'
    elif 15 <= hour < 18:
        return 'sore'
    else:
        return 'malam'

df['jam_published'] = df['published_local'].dt.hour.apply(kategori_jam)
```

```python
import pandas as pd

# ubah ke datetime
df['created'] = pd.to_datetime(df['created'], errors='coerce', utc=True)

# buat kolom kategori jam
# jika mau ke WIB (+7)
df['created_local'] = df['created'].dt.tz_convert('Asia/Jakarta')

def kategori_jam(hour):
    if 5 <= hour < 11:
        return 'pagi'
    elif 11 <= hour < 15:
        return 'siang'
    elif 15 <= hour < 18:
        return 'sore'
    else:
        return 'malam'

df['jam_created'] = df['created_local'].dt.hour.apply(kategori_jam)
```

## konversi boolean

```python
import pandas as pd

def split_known(x):
    date = pd.to_datetime(x, errors='coerce', utc=True)
    if pd.notna(date):
        return pd.Series({'known_by_police_date': date, 'known_by_police_source': None})
    else:
        return pd.Series({'known_by_police_date': None, 'known_by_police_source': x})

df[['known_by_police_date', 'known_by_police_source']] = df['known_by_police'].apply(split_known)
```

```python
import pandas as pd
import numpy as np

def split_es(x):
    val = str(x)
    date = pd.to_datetime(x, errors='coerce', utc=True)
    if pd.notna(date):
        return pd.Series({'published_to_es_flag': True, 'published_to_es_date': date})
    elif val.lower() == 'true':
        return pd.Series({'published_to_es_flag': True, 'published_to_es_date': np.nan})
    else:
        return pd.Series({'published_to_es_flag': False, 'published_to_es_date': np.nan})

df[['published_to_es_flag', 'published_to_es_date']] = df['published_to_es'].apply(split_es)
```

```python
import pandas as pd
import numpy as np

def split_is_updated(x):
    val = str(x)
    date = pd.to_datetime(x, errors='coerce', utc=True)
    if pd.notna(date):
        # kalau berupa tanggal valid
        return pd.Series({'is_updated_flag': True, 'is_updated_date': date})
    elif val.lower() == 'true':
        return pd.Series({'is_updated_flag': True, 'is_updated_date': np.nan})
    else:
        return pd.Series({'is_updated_flag': False, 'is_updated_date': np.nan})

df[['is_updated_flag', 'is_updated_date']] = df['is_updated'].apply(split_is_updated)
```

```python
import pandas as pd
import numpy as np

def split_published_all(x):
    date = pd.to_datetime(x, errors='coerce', utc=True)
    if pd.notna(date):
        return pd.Series({'published_all_date': date, 'published_all_source': None})
    else:
        return pd.Series({'published_all_date': None, 'published_all_source': x})

df[['published_all_date', 'published_all_source']] = df['published_all'].apply(split_published_all)
```
